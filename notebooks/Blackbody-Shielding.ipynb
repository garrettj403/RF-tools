# Black Body Shielding Required for Cryogenic Experiements

```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.constants as sc

# SciencePlots (optional)
plt.style.use(['science', 'notebook'])
```

```python
def shielding(freq, temp_bb, temp_exp):
    """Based on the equation in Glattli et al. 1997."""
    term1 = np.exp(sc.h * freq / sc.k / temp_bb) - 1
    term2 = np.exp(sc.h * freq / sc.k / temp_exp) - 1
    return 20 * np.log10(abs(term1 / term2))
```

```python
fig, ax = plt.subplots(figsize=(11, 7))

freq = np.logspace(6, 11, 301)

ax.semilogx(freq, shielding(freq, 300, 4.000), label="4 K")
ax.semilogx(freq, shielding(freq, 300, 0.250), label="250 mK")
ax.semilogx(freq, shielding(freq, 300, 0.100), label="100 mK")
ax.semilogx(freq, shielding(freq, 300, 0.050), label="50 mK")
ax.semilogx(freq, shielding(freq, 300, 0.015), label="15 mK")

plt.xlabel("Frequency")
plt.ylabel("Required attenuation (dB)")

plt.xlim([freq.min(), freq.max()])
plt.ylim([-499, 0])

plt.xticks([1e6, 1e7, 1e8, 1e9, 1e10, 1e11],
           ['1 MHz', '10 MHz', '100 MHz', '1 GHz', '10 GHz', '100 GHz'])

plt.legend(title="Base temperature", frameon=True, framealpha=1)
plt.title("Required shielding from 300 K reservoir")

plt.grid(which='both')
plt.tight_layout()

plt.savefig("figures/shielding-300k.png");
```
